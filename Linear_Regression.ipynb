```python
#Importing libraries

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#Importing the dataset

df=pd.read_csv('data/income.data.csv')
print(df.head())
```

```python
# Plotting the dataset by using scatter plot

plt.scatter(df.income,df.happiness,color="red",marker='*')
plt.xlabel('income')
plt.ylabel('happiness')
plt.show()
```

```python
# Defining X and Y

x= df['income'].values.reshape(-1,1) # -1,1 is to consider only income column
y= df['happiness'].values.reshape(-1,1) # consider only happiness column 
print(x)
# print(y)
```

```python
# Splitting the dataset into the Training set and Test set

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=0)  # 20% records would be used for testing..
print(X_train) # 389 records
#print(y_train) # 100 records

```

```python
# importing the model
from sklearn.linear_model import LinearRegression
lr = LinearRegression()

# fitting the model
lr.fit(X_train,y_train)

#Predict
y_pred =lr.predict(X_test)
print("predicted values:",y_pred)

#Printing Values
print('Values of Intercept:',lr.intercept_)
print('Value of Coefficient :',lr.coef_)
```

```python
# Model Evaluation

from sklearn import metrics
rmse = np.sqrt(metrics.mean_squared_error(y_test,y_pred))
R2 = metrics.r2_score(y_test,y_pred)  # r2_score always between 0 and 1
print("RMSE Score :",rmse)
print("R2_Score :",R2)
print("y_test values :",y_test)
print("y_pred values :",y_pred)

```

```python
#comparision of actual values and predicted values

dt =pd.DataFrame({'Actual' : y_test.flatten(), 'Predicted': y_pred.flatten()})
print(dt)
```
